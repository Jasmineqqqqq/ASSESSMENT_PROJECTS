```python
#KNN IMPUTER

df = pd.DataFrame({'Salary':[25000,48000, 71000, 85000, 90000, 49000],
                   'City':['Bengaluru', 'Delhi', 'Hyderabad', 'Bengaluru', 'Hyderabad', 'Bengaluru'],
                   'Gender':['Male', 'Female', 'Female', 'Female', 'Male', 'Male'],
                            'Exp':[1,3,5,6,9, None]})
df
```

```python
# Knn imputer will try to find the relation with other columns and impute the data according the relation with other columns
# In this case Age NaN is depending on the similarity with Fare columns
from sklearn.impute import KNNImputer
```

```python
knnimp = KNNImputer(n_neighbors=3)
knn_imp = pd.DataFrame(knnimp.fit_transform(df[['Salary', 'Exp']]))
knn_imp
#it takes the closest 3 neoghnors, the value for 49,000 salary is nan, so knn will ake average of 3 neighnours and impute nan vales
#so it will take 3 numbers closest to 49k in experience column, so 1,3 and 5, when u take their avrage, it will give 3
#hence nan column is now 3
```

```python
#ITERATIVE IMPUTER
#Before using Iterative Imputer, we need to enable it using below code
from sklearn.experimental import enable_iterative_imputer
#import Iterative Imputer
from sklearn.impute import IterativeImputer
```

```python
df2= pd.DataFrame({'Salary':[25000, 48000, 71000, 85000, 90000, 55000],
                   'City': [ 'Bengaluru', 'Delhi','Hyderabad', 'Bengaluru','Hyderabad', 'Bengaluru'],
                   'Gender': ['Male','Female', 'Female', 'Female', 'Male', 'Male'],
                   'Exp': [1,3,5,6,9, None]})
df2
```

```python
#This method treat other columns (which doesnot have nulls as feature and train on them and treat Null column
#as label). Finally it will predict the NaN data and impute. Its just like regression problem. Here Null column is
#label.
#here the nan column is exp, so it becomes the label and the salry column becomes the feature.
#it works like linear regression, it will learn the model to predict the nan value.. from previous data comparision
```

```python
iter_impute=IterativeImputer()
ite_imp = pd.DataFrame(iter_impute.fit_transform(df2[['Salary', 'Exp']]), columns=['Salary', 'Exp'])
ite_imp
```
